```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

two binomial distributions 1) with a p=0.2, and 2) with a p=0.5.

within the numpy binomial fxn 
<br>n = # of samples
<br>p = probability of success
<br>Note: the prob of success means the chance of 1 of 2 independent outcomes occuring
<br>in a sample. For the case of a fair coin it would be 0.5
<br>size = # of experiments

Going to increase the number of samples from 100 to 1000 based on question 1


```python
pop1 = np.random.binomial(1000,0.2,10000)
pop2 = np.random.binomial(1000,0.5,10000)

plt.hist(pop1, alpha=0.5, label='Population 1')
plt.hist(pop2, alpha=0.5, label='Population 2')
plt.legend(loc='upper right')
plt.show()
```

As the sample number increases the distributions drift away from eachother.

Next take a sample of the population creat a histogram, take the mean and the median.

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()
print(diff)
```

<br>The mean of the first sample is lower than the second as expected
<br>The std looks to be about the same. with sample1 lower
<br>Going to repeat this with a sample of 20 and compare

```python
pop1 = np.random.binomial(20,0.2,10000)
pop2 = np.random.binomial(20,0.5,10000)

plt.hist(pop1, alpha=0.5, label='Population 1')
plt.hist(pop2, alpha=0.5, label='Population 2')
plt.legend(loc='upper right')
plt.show()

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()
print(diff)
```

<br>Again the mean of sample1 is lower than sample2 but the overall
<br>magnitude of the number is lower as well
<br>std sample2 is greater than sample1 
<br>
<br>next going to change the p=0.3 and compute the t-static and pvalue.
<br>the distributions should get closer

```python
pop1 = np.random.binomial(20,0.3,10000)
pop2 = np.random.binomial(20,0.5,10000)

plt.hist(pop1, alpha=0.5, label='Population 1')
plt.hist(pop2, alpha=0.5, label='Population 2')
plt.legend(loc='upper right')
plt.show()

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()
print(diff)

from scipy.stats import ttest_ind
print(' ')
print('ttest results')
print(ttest_ind(sample2, sample1, equal_var=False))
```

The values appear to be closer than the p=0.2
<br>run this again with p=0.4

```python
pop1 = np.random.binomial(20,0.4,10000)
pop2 = np.random.binomial(20,0.5,10000)

plt.hist(pop1, alpha=0.5, label='Population 1')
plt.hist(pop2, alpha=0.5, label='Population 2')
plt.legend(loc='upper right')
plt.show()

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()
print(diff)

from scipy.stats import ttest_ind
print(' ')
print('ttest results')
print(ttest_ind(sample2, sample1, equal_var=False))
```

Again the distributions are getting closer because the probabilities
<br> are closer. The t-value went down as well which corresponds to a <br>greater overlap. The p-value also increased in size which is <br>indicative of the difference in population being a result of variance
<br>and not as much real difference
<br>
<br>next going to change the type of distrution to normal
<br>I am going to use a mean of 10 for sample1 and std 20
<br>sample2 will get a mean of 20 and std of 20

```python
pop1 = np.random.normal(10,20,10000)
pop2 = np.random.normal(20,20,10000)

plt.hist(pop1, alpha=0.5, label='Population 1')
plt.hist(pop2, alpha=0.5, label='Population 2')
plt.legend(loc='upper right')
plt.show()

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()
print(diff)

from scipy.stats import ttest_ind
print(' ')
print('ttest results')
print(ttest_ind(sample2, sample1, equal_var=False))
```

<br>the means and stds are obviously in line with what I entered
<br>the ttest shows tvalue of 3 which could be low to show overlap and
<br>pvalue of 0.001 which shows diff in pop as a result of variance
<br>going to run this using a wider spread in the means

```python
pop1 = np.random.normal(10,20,10000)
pop2 = np.random.normal(100,20,10000)

plt.hist(pop1, alpha=0.5, label='Population 1')
plt.hist(pop2, alpha=0.5, label='Population 2')
plt.legend(loc='upper right')
plt.show()

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()
print(diff)

from scipy.stats import ttest_ind
print(' ')
print('ttest results')
print(ttest_ind(sample2, sample1, equal_var=False))
```

<br>tvalue increased as the distributions drifted apart, pvalue down
<br>more - this is indicative of a higher variability sample as it is <br>more likely that the difference in the population meaningly reflects
<br>the population
