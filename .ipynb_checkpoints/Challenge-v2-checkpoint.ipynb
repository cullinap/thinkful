From our analysis above it looks like Population, GDP, and Hep B have the greatest amount of missing data so I will address those columns first. Let's look at the data to see how we can address it.

```python
missing_data = ['Population','Hepatitis B','GDP'] 

for column in missing_data:
    print("Unique values in column {} are: {}".format(column, lifeexpectancy_df[column].unique()))
```

```python
lifeexpectancy_df[pd.isnull(lifeexpectancy_df['Population'])].Country.unique()
```

The above list are countries that have missing population numbers. Since the dataset appears to span 16 years I thought it might be possible to carry some of the population data from a previous year into all of the other columns, but as you can see below from what it appears a country that is missing population data does not have any values associated with it. This looks like it generally holds for true for all these countries. There also seems to be a pattern with GDP. For the time being we could probably just ignore Population data and maybe GDP and use other values.

```python
lifeexpectancy_df[lifeexpectancy_df['Country']=='United States of America'].head()
```

```python
lifeexpectancy_df[pd.isnull(lifeexpectancy_df[' thinness  1-19 years'])].Country.unique()
```

```python
lifeexpectancy_df[lifeexpectancy_df['Country']=='United States of America']
```

It appears alcohol is just missing one year of data for most of the countries (2016). So it seems resonable to interpolate the data there. The same thing for Hep B as all of the countries have some Hep B scores.

```python
#interpolate alcohol
lifeexpectancy_df['Alcohol'] = lifeexpectancy_df['Alcohol'].interpolate()

#interpolate Hep B
lifeexpectancy_df['Hepatitis B'] = lifeexpectancy_df['Hepatitis B'].interpolate()
```

```python
#interpolate Total Expenditure
lifeexpectancy_df['Total expenditure'] = lifeexpectancy_df['Total expenditure'].interpolate()
```

```python
#drop countries with no population data
no_pop = lifeexpectancy_df[pd.isnull(lifeexpectancy_df['Population'])].Country.unique()

lifeexpectancy_data = lifeexpectancy_df[~lifeexpectancy_df.Country.isin(no_pop)]
```

```python
#drop all other NA's
```

```python
lifeexpectancy_data = lifeexpectancy_data.dropna()
```

### Outliers

Winsorize data

```python
for column in numerical_data:
    lifeexpectancy_data[column] = winsorize(lifeexpectancy_data[column])
```

```python
plt.figure(figsize = (20,40))

for i in range(19):
    plt.subplot(7,3,i+1)
    sns.distplot(lifeexpectancy_data[numerical_data[i]].dropna())
    plt.title(numerical_data[i])

plt.show()
```

```python
plt.figure(figsize = (20,40))

for i in range(19):
    plt.subplot(7,3,i+1)
    sns.boxplot(lifeexpectancy_data[numerical_data[i]])
    plt.title(numerical_data[i])

plt.show()
```

#### Jarque-Bera and Normal tests

```python
for i in range(19):  
    jb_stats = jarque_bera(np.sqrt(lifeexpectancy_data[numerical_data[i]]))
    norm_stats = normaltest(np.sqrt(lifeexpectancy_data[numerical_data[i]]))

    print("Jarque-Bera test statistics is {0} and p value is {1}".format(jb_stats[0], jb_stats[1]))
    print("Normality test statistics is {0} and p value is {1}".format(norm_stats[0], norm_stats[1]))
```

### PCA



```python
corrmat_student = lifeexpectancy_data.corr()

plt.figure(figsize=(20,10))

sns.heatmap(corrmat_student, square=True, annot=True, linewidths=.5)
plt.title("correlation matrix (life expectancy)")

plt.show()
```

```python
X = lifeexpectancy_data[numerical_data]

X = StandardScaler().fit_transform(X)

sklearn_pca = PCA(n_components=1)
lifeexpectancy_data['pca_1'] = sklearn_pca.fit_transform(X)

print(
    'The percentage of total variance in the dataset explained by each',
    'component from Sklearn PCA.\n',
    sklearn_pca.explained_variance_ratio_
)
```

```python
corrmat_student = lifeexpectancy_data[['pca_1','Life expectancy ']].corr()

plt.figure(figsize=(20,5))

sns.heatmap(corrmat_student, square=True, annot=True, linewidths=.5)
plt.title("correlation matrix (student performance)")

plt.show()
```
