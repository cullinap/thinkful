Choose a state (e.g. California) and draw a line graph of its total revenues and total expenditures along the years. How do these 2 variables evolve during the years? Try to explain the peaks and troughs with some state-wise news and information around those dates.

```python
new_york_data = useducation_df[useducation_df['STATE']=='NEW_YORK']

fig, ax = plt.subplots(figsize=(20,8))

sns.lineplot(x='YEAR',y='TOTAL_REVENUE',data=new_york_data)
sns.lineplot(x='YEAR',y='TOTAL_EXPENDITURE',data=new_york_data)
plt.title('Total Revenue v. Total Expenditure in New York State')
ax.set_ylabel('Money')
tot_rev = mpatches.Patch(color='blue', label='TOTAL_REVENUE')
tot_exp = mpatches.Patch(color='orange', label='TOTAL_EXPENDITURE')
plt.legend(title='Legend', loc='best', prop={'size': 10}, handles=[tot_rev,tot_exp])

plt.show()
```

### Question 4

What are the distributions of the math and reading scores in the sample?

```python
plt.figure(figsize = (20,5))

plt.subplot(1,4,1)
sns.distplot(new_york_data['AVG_MATH_4_SCORE'].dropna())
plt.title('NYS AVG_MATH_4_SCORE')

plt.subplot(1,4,2)
sns.distplot(new_york_data['AVG_MATH_8_SCORE'].dropna())
plt.title('NYS AVG_MATH_8_SCORE')

plt.subplot(1,4,3)
sns.distplot(new_york_data['AVG_READING_4_SCORE'].dropna())
plt.title('NYS AVG_READING_4_SCORE')

plt.subplot(1,4,4)
sns.distplot(new_york_data['AVG_READING_8_SCORE'].dropna())
plt.title('NYS AVG_READING_8_SCORE')

plt.show()
```

The data shows a median that is higher than the mean with a longer tail on the left. This could be due to the fact that the numbers approach a max test score on the right side.

### Question 4

Notice there are too many missing values for math and reading scores. Fill out the missing values using mean, median, and linear interpolation. Then compare the effects of these techniques on the distributions of the score variables.

```python
test_results = columns[21:25]

na_value_counts = [new_york_data[column].isnull().sum() for column in test_results]
tot_value_counts = [new_york_data[column].count() for column in test_results]

for test,value,count in zip(test_results,na_value_counts,tot_value_counts):
    print(str(test) + (' ') + str(value) + (' ') + str(round((count/(count+value)*100),ndigits=1)))
```

#### Mean

```python
new_york_data_mean = new_york_data.copy()

[new_york_data_mean[column].fillna(new_york_data_mean[column].mean(),inplace=True) for column in test_results]
```

```python
plt.figure(figsize = (20,5))

plt.subplot(1,4,1)
sns.distplot(new_york_data_mean['AVG_MATH_4_SCORE'].dropna())
plt.title('NYS AVG_MATH_4_SCORE')

plt.subplot(1,4,2)
sns.distplot(new_york_data_mean['AVG_MATH_8_SCORE'].dropna())
plt.title('NYS AVG_MATH_8_SCORE')

plt.subplot(1,4,3)
sns.distplot(new_york_data_mean['AVG_READING_4_SCORE'].dropna())
plt.title('NYS AVG_READING_4_SCORE')

plt.subplot(1,4,4)
sns.distplot(new_york_data_mean['AVG_READING_8_SCORE'].dropna())
plt.title('NYS AVG_READING_8_SCORE')

plt.show()
```

#### Median

```python
new_york_data_median = new_york_data.copy()

[new_york_data_median[column].fillna(new_york_data_median[column].median(),inplace=True) for column in test_results]
```

```python
plt.figure(figsize = (20,5))

plt.subplot(1,4,1)
sns.distplot(new_york_data_median['AVG_MATH_4_SCORE'].dropna())
plt.title('NYS AVG_MATH_4_SCORE')

plt.subplot(1,4,2)
sns.distplot(new_york_data_median['AVG_MATH_8_SCORE'].dropna())
plt.title('NYS AVG_MATH_8_SCORE')

plt.subplot(1,4,3)
sns.distplot(new_york_data_median['AVG_READING_4_SCORE'].dropna())
plt.title('NYS AVG_READING_4_SCORE')

plt.subplot(1,4,4)
sns.distplot(new_york_data_median['AVG_READING_8_SCORE'].dropna())
plt.title('NYS AVG_READING_8_SCORE')

plt.show()
```

```python
education_df = new_york_data.copy()

fill_list = ["ENROLL", "TOTAL_REVENUE", "FEDERAL_REVENUE",
             "STATE_REVENUE", "LOCAL_REVENUE", "TOTAL_EXPENDITURE",
             "INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
             "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE", "GRADES_PK_G",
             "GRADES_KG_G", "GRADES_4_G", "GRADES_8_G", "GRADES_12_G", "GRADES_1_8_G",
             "GRADES_9_12_G", "GRADES_ALL_G"]

states = education_df["STATE"].unique()

for state in states:
    education_df.loc[education_df["STATE"] == state, fill_list] = education_df.loc[education_df["STATE"] == state, fill_list].interpolate()

# we drop the null values after interpolation
education_df.dropna(inplace=True)
```

```python
plt.figure(figsize = (20,5))

plt.subplot(1,4,1)
sns.distplot(education_df['AVG_MATH_4_SCORE'].dropna())
plt.title('NYS AVG_MATH_4_SCORE')

plt.subplot(1,4,2)
sns.distplot(education_df['AVG_MATH_8_SCORE'].dropna())
plt.title('NYS AVG_MATH_8_SCORE')

plt.subplot(1,4,3)
sns.distplot(education_df['AVG_READING_4_SCORE'].dropna())
plt.title('NYS AVG_READING_4_SCORE')

plt.subplot(1,4,4)
sns.distplot(education_df['AVG_READING_8_SCORE'].dropna())
plt.title('NYS AVG_READING_8_SCORE')

plt.show()
```
