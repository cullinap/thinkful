```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
coin_flips = []

def coin_toss(tosses):
    
    for i in range(tosses):
        result = np.random.randint(2)
        coin_flips.append(result)
    
    return coin_flips
```

```python
#heads = 1, tails = 0
sns.countplot(coin_toss(100000))
plt.show()
```

```python
def bet_coin(money,flips):
    money = money

    for i in range(flips):
        result = np.random.randint(2)
        if result == 0:
            money = money - 100
        else:
            money = money + 50
        return money
```

```python
earnings = []

for i in range(100000):
    x = bet_coin(1000,100)
    earnings.append(x)
```

```python
sns.distplot(earnings)
plt.show()
```
